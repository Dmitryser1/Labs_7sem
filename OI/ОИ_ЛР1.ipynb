```python
## 1-2 task
import numpy as np
import matplotlib.pyplot as plt

# Начальное значение параметра B
beta = 1/10

# Число точек
n = 1000

# Значения параметра B, которые мы хотим исследовать
B_values = [beta, 0.1 * beta, -beta]

# Функция для вычисления сигнала f(x) для данного B
def signal(beta, x):
    return np.exp(1j * beta * x)

# Создаем массив значений x
x = np.linspace(0, 2 * np.pi, n)

# Создаем подграфики для разных значений B
fig, axs = plt.subplots(len(B_values), 2, figsize=(10, 8))

# Названия для графиков
titles = ['Amplitude', 'Phase']

for i, B in enumerate(B_values):
    # Вычисляем сигнал для данного B
    f_x = signal(B, x)
    
    # Выводим амплитуду на первом подграфике
    axs[i, 0].plot(x, np.abs(f_x))
    axs[i, 0].set_title(titles[0] + f' (B={B})')
    
    # Выводим фазу на втором подграфике
    axs[i, 1].plot(x, np.angle(f_x))
    axs[i, 1].set_title(titles[1] + f' (B={B})')

plt.tight_layout()
plt.show()
```
