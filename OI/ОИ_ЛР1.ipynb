```python
## 1-2 task
import numpy as np
import matplotlib.pyplot as plt

#    B
beta = 1/10

#  
n = 1000

#   B,    
B_values = [beta, 0.1 * beta, -beta]

#     f(x)   B
def signal(beta, x):
    return np.exp(1j * beta * x)

#    x
x = np.linspace(0, 2 * np.pi, n)

#      B
fig, axs = plt.subplots(len(B_values), 2, figsize=(10, 8))

#   
titles = ['Amplitude', 'Phase']

for i, B in enumerate(B_values):
    #     B
    f_x = signal(B, x)
    
    #     
    axs[i, 0].plot(x, np.abs(f_x))
    axs[i, 0].set_title(titles[0] + f' (B={B})')
    
    #     
    axs[i, 1].plot(x, np.angle(f_x))
    axs[i, 1].set_title(titles[1] + f' (B={B})')

plt.tight_layout()
plt.show()
```
